```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
```

```python
from nltk.corpus import stopwords
import string
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
import nltk
nltk.download('stopwords')
```

```python
stop = set(stopwords.words('english'))
exclude = set(string.punctuation)
lemma = WordNetLemmatizer()

def clean(issue):
    stop_free = ' '.join([word for word in issue.lower().split() if word not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = ' '.join([lemma.lemmatize(word) for word in punc_free.split()])
    return normalized.split()
```

```python
# converting the given text term-document matrix

vectorizer = TfidfVectorizer(max_features=1500, min_df=10, stop_words='english')
X = vectorizer.fit_transform(text_data)
words = np.array(vectorizer.get_feature_names_out())

print(X)
print("X = ", words)
```

```python
# Applying Non-Negative Matrix Factorization

nmf = NMF(n_components=8, solver="mu")
W = nmf.fit_transform(X)
H = nmf.components_

for i, topic in enumerate(H):
     print("Topic {}: {}".format(i + 1, ",".join([str(x) for x in words[topic.argsort()[-8:]]])))
```

```python
topic_arr = []
for i, topic in enumerate(H):
 topic_arr.append([str(x) for x in words[topic.argsort()[-8:]]])
```

```python
df_topic = pd.DataFrame([])
df_topic["category"] = topic_arr
```
